```python
class Food(object):
    def __init__(self, name, price, description):
        self.name = name
        self.price = price
        self.description = description
    
    def getprice(self):
        return self.price
    
    def __str__(self):
        return self.name + ' \t\t ' + str(self.getprice()) + " \t " + self.description

menu = []

while True:
    print("\nChoose an option\n 1. Add food item to menu \n 2. Display menu \n 3. search a food item\n 4. Edit a food item from the menu\n 5. Delete an item from menu \n 6. exit application\n")
    try:
        selection = int(input("Enter your option : "))
        if(selection == 1):
            item = input("Enter name of the item : ")
            if item in [ ele for ele in menu if ele.name == item]:
                print("This food item already exists in the menu!")
            else:
                while True:
                    try:
                        price = int(input("Enter price of the item : "))
                        if price > 0:
                            break   
                        else:
                            print("Please enter positive value: ")                         
                    except ValueError:
                        print("Please enter correct amount")
                description = input("Enter description of the item : ")
                menu.append(Food(item, price, description))
                print(f"{item} is added to the menu")
            
        elif(selection == 2):
            if len(menu) != 0:
                print("name\t\t\t\tprice\t\t\t\tdescription")
                print("----\t\t\t\t-----\t\t\t\t-----------")
                for el in menu:
                    print(el.name + "\t\t\t\t" + str(el.price) + "\t\t\t\t" + el.description)
            else:
                print("No Items added yet !\n")
            
        elif selection == 3:
            item = input("Enter item to search: ")
            res_item = [ele for ele in menu if ele.name == item]
            if len(res_item) != 0:
                print("\nResult :")
                for el in res_item:
                    print(el.name + "\t\t" + str(el.price) + "\t\t" + el.description)
            else:
                print("No items found!")
        elif selection == 4:
            item = input("Enter item to edit: ")
            res_item = [ele for ele in menu if ele.name == item]
            if len(res_item) != 0:
                print("Item saved in menu as :")
                for el in res_item:
                    print(el.name + "\t\t" + str(el.price) + "\t\t" + el.description)
                print("Select an option \n1. Change price of the item \n2. Change description of the item\n")
                sel = int(input("Enter the option to change the item :"))
                if sel == 1:
                    while True:
                        try:
                            price = int(input("Enter latest price: "))
                            if(price > 0):
                                break
                            else:
                                print("Please enter positive value: ")
                            break
                        except ValueError:
                            print("Please enter correct amount")
                    for e in res_item:
                        e.price = price
                    print("Item price updated in menu")
                elif sel == 2:
                    description = input("Enter latest price: ")
                    for e in res_item:
                        e.description = description
                    print("Item description updated in menu")
            else:
                print("\n-----No items found!-----")
        elif selection == 5:
            item = input("Enter item to delete: ")
            res_item = [ele for ele in menu if ele.name == item]
            if len(res_item) != 0:
                print("Item found to delete :")
                for el in res_item:
                    print(el.name + "\t\t\t\t" + str(el.price) + "\t\t\t\t" + el.description)
                response = input("\nDo you want to delete this item (Y/N) : ")
                if(response[0].upper() == 'Y'):
                    for el in res_item:
                        menu.remove(el)
                    print(f"{item} is deleted")
                else:
                    print("----No items deleted----")
            else:
                print("\n-----No items found to delete!-----")
        else:
            break

    except ValueError:
        print("Enter correct option")
```
