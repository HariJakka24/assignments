```python
class Student():
    def __init__(self, name, height):
        self.name = name
        self.height = height

records = []

while True:
    print("\nChoose an option\n 1. Add student height to record \n 2. Display heights \n 3. search for height of a student\n 4. Edit height of a student\n 5. Delete a student's record \n 6. exit application\n")
    try:
        selection = int(input("Enter your option : "))
        if(selection == 1):
            name = input("Enter name of the student : ")
            if name in [ ele for ele in records if ele.name == name]:
                print("This Student already exists in the records!")
            else:
                while True:
                    try:
                        height = float(input(f"Enter height of the {name} in cm : "))
                        if height > 0:
                            break   
                        else:
                            print("Height should be a positive value: ")                         
                    except ValueError:
                        print("Please enter correct value")
                records.append(Student(name, height))
                print(f"{name}'s height is recorded")
            
        elif(selection == 2):
            if len(records) != 0:
                print("name\t\t\t\t\theight")
                print("----\t\t\t\t\t-----")
                for el in records:
                    print(el.name + "\t\t\t\t\t" + str(el.height))
            else:
                print("No Students added yet !\n")
            
        elif selection == 3:
            name = input("Enter student's name to search: ")
            res_name = [ele for ele in records if ele.name == name]
            if len(res_name) != 0:
                print("\nResult :")
                for el in res_name:
                    print(el.name + "\t\t\t\t" + str(el.height))
            else:
                print("No names found!")
        elif selection == 4:
            name = input("Enter name to edit: ")
            res_name = [ele for ele in records if ele.name == name]
            if len(res_name) != 0:
                print(f"previous height of {name} is :")
                for el in res_name:
                    print(el.name + "\t\t\t\t" + str(el.height))
                while True:
                        try:
                            height = float(input("Enter latest height: "))
                            if(height > 0):
                                break
                            else:
                                print("Please enter positive value: ")
                            break
                        except ValueError:
                            print("Please enter correct amount")
                for e in res_name:
                    e.height = height
                print(f"{name}'s height is updated in records")
            else:
                print("\n-----No names found!-----")
        elif selection == 5:
            name = input("Enter student's name to delete: ")
            res_name = [ele for ele in records if ele.name == name]
            if len(res_name) != 0:
                print("student found to delete :")
                for el in res_name:
                    print(el.name + "\t\t\t\t" + str(el.height))
                response = input("\nDo you want to delete this name (Y/N) : ")
                if(response[0].upper() == 'Y'):
                    for el in res_name:
                        records.remove(el)
                    print(f"{name}'s record is deleted")
                else:
                    print("----No record is deleted----")
            else:
                print("\n-----No student found to delete!-----")
        elif selection == 6:
            break
        else:
            print("Please enter correct option from the list")

    except ValueError:
        print("Enter correct option")
```
